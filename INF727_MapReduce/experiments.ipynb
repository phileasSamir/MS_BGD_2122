```python
import numpy as np
import subprocess
import os
import glob
import time
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import cm 

for _ in range(1):
    print("big")
    numbas = [1,2,3,4,6,12]
    for x, y in list(zip(numbas, numbas[::-1]))[2:]: #1x12, 2x6 will not succeed with our implementation due to memory issues
        t0 = time.time()
        out, err = subprocess.Popen(f"python .\MASTER_BIG.py .\machines_full.txt {x} {y}".split(),stdout=subprocess.PIPE).communicate()
        t1 = time.time()
        open("logs/big_logs.log","a",encoding="utf8").write(out.decode("utf8"))
        print(f"machines : {x}, splits : {y}, time : {int(t1-t0)}s")
    print("bigger")
    numbas = [1,2,3,4,6,8,12,24]
    for x, y in list(zip(numbas, numbas[::-1]))[4:]: #1x24, 2x12, 3x8, 4x6 will not succeed with our implementation due to memory issues
        t0 = time.time()
        out, err = subprocess.Popen(f"python .\MASTER_BIG.py .\machines_full.txt {x} {y}".split(),stdout=subprocess.PIPE).communicate()
        t1 = time.time()
        open("logs/big_logs.log","a",encoding="utf8").write(out.decode("utf8"))
        print(f"machines : {x}, splits : {y}, time : {int(t1-t0)}s")
    for file in glob.glob("data/*"):
        print(file)
        machines, splits = np.meshgrid([i for i in range(1,21)],[i for i in range(1,21)])
        for x, y in zip(machines.flatten(),splits.flatten()):
            t0 = time.time()
            out, err = subprocess.Popen(f"python .\MASTER.py .\machines_full.txt {x} {y} {file}".split(),stdout=subprocess.PIPE).communicate()
            t1 = time.time()
            open("logs/mapreduce_logs.log","a",encoding="utf8").write(out.decode("utf8"))
            print(f"machines : {x}, splits : {y}, time : {int(t1-t0)}s")
```
