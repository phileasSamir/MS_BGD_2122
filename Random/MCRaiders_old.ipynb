```python
def guess_year(page_number, month):
    page_number = int(page_number)
    month = int(month)
    if page_number == 11:
        if int(month) > 10:
            year = 2020
        else:
            year = 2021
    elif page_number > 21:
        if int(month) > 6:
            year = 2019
        else:
            year=2020
    elif page_number > 11:
        year = 2020
    elif page_number < 5 and int(month) == 1:
        year = 2022
    else:
        year = 2021
    return year

newcols = list()
for col in df.columns:
    newcols.append(col.split("_")[0][3:]+"/"+str(guess_year(col.split("_")[-1], col.split("_")[0].split("/")[-1])))

print(len(newcols))
newcols
```
