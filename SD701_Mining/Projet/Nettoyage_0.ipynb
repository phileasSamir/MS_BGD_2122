# Nettoyage fichier dep20

```python
dep20 = pd.read_excel("DEP20.xlsx")
```

```python
dep20 = dep20.drop(columns=['Libellé du département', 'Code du canton',
       'Libellé du canton', 'Code de la commune', 'Libellé de la commune',
       'Inscrits', '% Abs/Ins', '% Vot/Ins', '% Blancs/Ins', '% Blancs/Vot', '% Nuls/Ins',
       '% Nuls/Vot', 'Exprimés', '% Exp/Ins', '% Exp/Vot'])
```

```python
#On additionne les votes nuls, blancs, et les abstentions
dep20["ABN"]=dep20['Abstentions']+dep20["Blancs"]+dep20["Nuls"]
dep20 = dep20.drop(columns=["Abstentions","Blancs","Nuls"])
```

```python
dep20col = dep20.columns
dep20col
```

```python
#pour les colonnes de scrutin, on ne garde que les "nuance" et "voix"
#liste des colonnes à supprimer
liste = list(dep20col[2:-1:6])+list(dep20col[3:-1:6])+list(dep20col[6:-1:6])+list(dep20col[7:-1:6])
print(liste)
dep20.drop(columns=liste,inplace=True)
```

```python
dep20.head()
```

```python
dep20.columns
```

```python
dep20col = dep20.columns

#on renomme les colonnes 'panneau' "nuance" "voix" pour les distinguer
nuances_voix20 = []
for i in range(2, len(dep20col)-1, 2):
    nuances_voix20 += ['Nuance'+str(int((i-2)/2)), 
                        'Voix'+str(int((i-2)/2))]
#on récupère séparément les colonnes de nuances et de voix
nuances = nuances_voix20[0:-1:2]
voix = nuances_voix20[1::2]
    
dep20.columns = ['Code du département', 'Votants'] + nuances_voix20 + ['ABN']
#On réarrange les colonnes
dep20 = dep20[['Code du département', 'Votants', 'ABN'] + nuances_voix20]
```

```python
dep20.head()
```

```python
#on enlève les "BC-" devant les noms de partis
for nuance in nuances:
    dep20[nuance] = dep20[nuance].str.extract(r"BC-(.*)")
```

```python
#on remplace les noms de partis par leur mouvance politique
dep20 = dep20.replace(partis)
dep20.head()
```

```python
#agrégats par département des nb de votants et nb ABN
#dep20_2 = 
dep20_2 = dep20[["Code du département","Votants","ABN"]]
dep20_2 = dep20_2.groupby(by=['Code du département']).sum()
dep20_2.head()
```

```python
#par département, nb de votes par groupe de parti politique

dep20_3 = dep20.drop(columns=["Votants","ABN"])
dep20_3 = pd.melt(dep20_3, id_vars=['Code du département']+nuances,
                  value_vars=voix, var_name="Voix", value_name="Nb_votes")
dep20_3 = pd.melt(dep20_3, id_vars=['Code du département','Voix','Nb_votes'],
                  value_vars=nuances, var_name="nuance", value_name="parti")

#suppression des lignes pour lesquelles le nb de votes ou le parti = NaN
dep20_3.dropna(subset=['parti'], inplace=True)
dep20_3.dropna(subset=['Nb_votes'], inplace=True)

#Suppression des lignes pour lesquelles on a nuance(i),voix(j) avec i!=j
dep20_3.drop(dep20_3.loc[(dep20_3['Voix'].str.extract(r"([0-9]+)") != 
                          dep20_3['nuance'].str.extract(r"([0-9]+)")).values].index,
            inplace=True)

#Comme les nuances/voix correspondent désormais, on peut supprimer ces colonnes inutiles
dep20_3.drop(columns=["Voix","nuance"],inplace=True)
```

```python
dep20_3['parti'].unique()
```

```python
dep20_4 = dep20_3.groupby(by=['Code du département','parti']).sum().unstack(level=-1)
dep20_4 = dep20_4.droplevel(0, axis=1)
dep20_4.head()
```

```python
dep20_5 = dep20_2.join(dep20_4)
```

```python
dep20_5.rename(index={"ZA": 971, "ZD": 974, "ZM": 976},inplace=True)
```

```python
dep20_5 = dep20_5[['Votants', 'ABN', 'EG', 'G', 'C', 'D', 'ED', 'A']]
dep20_5.to_csv('dep20_clean.csv')
```
