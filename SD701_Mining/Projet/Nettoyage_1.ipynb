```python
import pandas as pd
```

## Nettoyage de EURO2019

```python
df_euro_2019 = pd.read_csv("RawData/europeennes_2019.txt",sep="\n",header=None,encoding="latin1")
df_euro_2019 = df_euro_2019[0].str.split(";", expand=True)
df_euro_2019.columns = df_euro_2019.loc[0]
df_euro_2019 = df_euro_2019.drop(0,axis=0)
df_euro_2019.head()
```

```python
l = df_euro_2019.columns.to_list()[:25] + df_euro_2019.columns.to_list()[18:25] * 33
df_euro_2019.columns = l

df_euro_2019 = df_euro_2019.drop(columns=['Libellé du département', 
                                          'Code de la commune', 'Libellé de la commune', 
                                          'Inscrits', '% Abs/Ins', '% Vot/Ins', '% Blancs/Ins', '% Blancs/Vot', '% Nuls/Ins', '% Nuls/Vot', 
                                          'Exprimés', '% Exp/Ins','% Exp/Vot',
                                          'N°Liste', 'Libellé Etendu Liste', 'Nom Tête de Liste', 
                                          '% Voix/Ins', '% Voix/Exp'])
df_euro_2019 = df_euro_2019.dropna(axis='columns',how='all')
```

```python
#On additionne les votes nuls, blancs, et les abstentions
df_euro_2019["ABN"]=df_euro_2019['Abstentions'].astype(int)+df_euro_2019["Blancs"].astype(int)+df_euro_2019["Nuls"].astype(int)
df_euro_2019 = df_euro_2019.drop(columns=["Abstentions","Blancs","Nuls"])
```

```python
df_euro_2019["Code du département"] = df_euro_2019["Code du département"].astype(str)
df_euro_2019 = df_euro_2019.replace({"2A": 20, "2B": 20,
                                     "ZA": 971, "ZB": 972, "ZC": 973, "ZD": 974, "ZM": 976, "ZN": 988, "ZP": 987, "ZS": 975, 
                                     "ZW": 986, "ZX": 977, "ZZ": 99  })
df_euro_2019["Code du département"] = df_euro_2019["Code du département"].astype(int)
df_euro_2019["Code du département"].unique()
```

```python
df_euro_2019.head()
```

```python
# valeur unique dans les colonnes "Libellé Abrégé Liste" ?
for i in range(2,70,2):
    if df_euro_2019.iloc[:,i].nunique() != 1:
        print("value not unique")
```

```python
grp_partis = {"EG":["DÉMOCRATIE REPRÉSENTATIVE", "DÉCROISSANCE 2019", "LUTTE OUVRIÈRE", "POUR L'EUROPE DES GENS", "RÉVOLUTIONNAIRE"], 
              "G":["LA FRANCE INSOUMISE", "URGENCE ÉCOLOGIE", "ENVIE D'EUROPE", "LISTE CITOYENNE", "ESPERANTO", "EUROPE ÉCOLOGIE"] , 
              "C":["RENAISSANCE", "LES EUROPÉENS"], 
              "D":["DEBOUT LA FRANCE", "ENSEMBLE POUR LE FREXIT", "UNION DROITE-CENTRE", "UDLEF"],
              "ED":["UNE FRANCE ROYALE", "LA LIGNE CLAIRE", "ENSEMBLE PATRIOTES", "LISTE DE LA RECONQUÊTE", "PRENEZ LE POUVOIR"] , 
              "A":["PARTI PIRATE", "PACE", "PARTI FED. EUROPÉEN", "INITIATIVE CITOYENNE", "ALLONS ENFANTS", "À VOIX ÉGALES", 
                   "NEUTRE ET ACTIF", "ÉVOLUTION CITOYENNE", "ALLIANCE JAUNE", "PARTI ANIMALISTE", "LES OUBLIES DE L'EUROPE",  
                   "EUROPE AU SERVICE PEUPLES"]}
partis = {keyi: value for (value, key) in grp_partis.items() for keyi in key}
```

```python
df_euro_2019 = df_euro_2019.replace(partis)
```

```python
EG = []
G = []
C = []
D = []
ED = []
A = []
for i in range(2,70,2):
    if df_euro_2019.iloc[0,i] == "EG":
        EG.append(i+1)
    elif df_euro_2019.iloc[0,i] == "G":
        G.append(i+1)
    elif df_euro_2019.iloc[0,i] == "C":
        C.append(i+1)
    elif df_euro_2019.iloc[0,i] == "D":
        D.append(i+1)
    elif df_euro_2019.iloc[0,i] == "ED":
        ED.append(i+1)
    elif df_euro_2019.iloc[0,i] == "A":
        A.append(i+1)
```

```python
df_euro_2019["EG"]=df_euro_2019.iloc[:,EG].astype(int).sum(axis=1)
df_euro_2019["G"]=df_euro_2019.iloc[:,G].astype(int).sum(axis=1)
df_euro_2019["C"]=df_euro_2019.iloc[:,C].astype(int).sum(axis=1)
df_euro_2019["D"]=df_euro_2019.iloc[:,D].astype(int).sum(axis=1)
df_euro_2019["ED"]=df_euro_2019.iloc[:,ED].astype(int).sum(axis=1)
df_euro_2019["A"]=df_euro_2019.iloc[:,A].astype(int).sum(axis=1)
```

```python
df_euro_2019["Votants"]=df_euro_2019["Votants"].astype(int)
```

```python
df_euro_2019 = df_euro_2019[["Code du département","Votants","ABN","EG","G","C","D","ED","A"]]
```

```python
df_euro_2019 = df_euro_2019.groupby("Code du département").sum()
```

```python
df_euro_2019
```

```python
df_euro_2019.to_csv("data_clean/euro2019_clean.csv")
```
