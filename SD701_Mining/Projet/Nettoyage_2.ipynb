```python
df = df_1000_plus#.append(df_1000_moins)
df.head()
```

```python
df["Code Nuance"] = df["Code Nuance"].str.extract(r"L(.*)")
```

```python
df["Code du département"] = df["Code du département"].astype(str)
df = df.replace({"ZA": "971", 
                 "ZB": "972", 
                 "ZC": "973", 
                 "ZD": "974", 
                 "ZS": "975", 
                 "ZP": "987", 
                 "ZN": "988", 
                 "ZM": "976", #"2A": 20, "2B": 20,
                 "ZW": "986",
                 "ZX": "978",
                 "ZZ": "99",
                 "1":"01",
                 "2":"02",
                 "3":"03",
                 "4":"04",
                 "5":"05",
                 "6":"06",
                 "7":"07",
                 "8":"08",
                 "9":"09"})
#df["Code du département"] = df["Code du département"].astype(int)
df["Code du département"].unique()
```

```python
grp_partis = {"EG": ['COM', 'EXG','FG','LO','NPA',"EG"], 
              "G":["SOC","DVG","PG", "UG","RDG","VEC", "LFI","FI","RDG","UGE","ECO","GJ","G"] , 
              "C":["MMD","CMD","MDM","UC","UDI","LREM","UCD","UCG","REM","DVC","CEN","NCE","ALLI","C"], 
              "D":["DVD","UD","UMP","LR","DSV","PRV","D"],
              "ED":["EXD","FN","RN","DLF","UXD","ED"] , 
              "A":["REG","AUT","DIV","A"]}
partis = {keyi: value for (value, key) in grp_partis.items() for keyi in key}

df = df.replace(partis)
df["Code Nuance"].value_counts()
```

```python
#df = df.groupby(["Code du département","Inscrits","Abstentions","Votants","Blancs et nuls"]).first()
df = df.pivot_table(values="Voix",columns="Code Nuance",index=["Code du département","Inscrits","Abstentions","Votants","Blancs et nuls"], aggfunc="sum", fill_value=0)
df = df.reset_index()
df.columns.name=""
df.head()
```

```python
df = df.groupby("Code du département").sum()
df.head()
```

```python
df["ABN"] = df["Abstentions"] + df["Blancs et nuls"]
df = df.drop(["Inscrits","Abstentions","Blancs et nuls"], axis=1)
df.head()
```

```python
df.to_csv("data_clean/muni2014clean.csv")
```
