```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import pygal.maps.fr
import glob

def distortionelbow(df,label,krange=range(2,10),abn=True):
    df = df.drop(["Votants"],axis=1)
    if not abn:
        df = df.drop(["ABN"],axis=1)
    X = df.values
    distortions = []
    for k in krange:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(X)
        distortions.append(kmeanModel.inertia_)
    #plt.figure(figsize=(8,6))
    sns.lineplot(x=krange,y=distortions, label=label)
    sns.scatterplot(x=krange,y=distortions)

def kmeans_elections(df, title, k=4, abn=True):
    df = df.drop(["Votants"],axis=1)
    if not abn:
        df = df.drop(["ABN"],axis=1)

    X = df.values

    length = np.sqrt((X**2).sum(axis=1))[:,None]
    X = X/length

    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    ccdf = pd.DataFrame(kmeans.cluster_centers_)
    ccdf.columns = df.columns
    display(ccdf.head())
    ccdf.to_latex(f"tex/{title}_centroids.tex")

    X_TSNE = TSNE().fit_transform(X)
    X_PCA = PCA(2).fit_transform(X)

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
    dimred = ["TSNE","PCA"]
    for i, X in enumerate([X_TSNE, X_PCA]):
        sns.scatterplot(x=X[:,0],y=X[:,1],hue=kmeans.labels_,palette="Set1",ax=ax[i]).set_title(dimred[i])
    plt.legend()
    plt.suptitle(title)
    plt.savefig(f"figs/{title}_dr.png")
    plt.show()
    df["labels"] = kmeans.labels_

    if 20 in df.index.to_list():
        display(df.loc[20,:])
        df1 = df.copy()
        df1 = pd.DataFrame([df1.loc[20,:].values,df1.loc[20,:].values], columns=df1.columns,index=["2A","2B"])
        display(df1.tail())
        df = df.append(df1)
        display(df.tail())

    fr_chart = pygal.maps.fr.Departments(human_readable=True,show_legend=False)
    fr_chart.title = f'KMeans labels, {title}'
    for i in range(len(df["labels"].unique())):
        fr_chart.add(f'Cluster {i}', df["labels"][df["labels"]==i].to_dict())
    fr_chart.render_to_png(f"figs/{title}_map.png")

    return df["labels"]

lections = ["Dep_15","Dep_21","Eur_14","Eur_19","Leg_12","Leg_17","Muni_14","Muni_20","Pdt_12","Pdt_17","Reg_15","Reg_21"]
fulldf = list()

plt.figure(figsize=(8,6))
for i, f in enumerate(glob.glob("./data_clean/*")):
    if i>=0:
        df = pd.read_csv(f, index_col=0).fillna(0)
        df.index = df.index.to_series().replace({"ZA": "971", 
                                                "ZB": "972", 
                                                "ZC": "973", 
                                                "ZD": "974", 
                                                "ZS": "975", 
                                                "ZP": "987", 
                                                "ZN": "988", 
                                                "ZM": "976", #"2A": 20, "2B": 20,
                                                "ZW": "986",
                                                "ZX": "978",
                                                "ZZ": "99",
                                                "1":"01",
                                                "2":"02",
                                                "3":"03",
                                                "4":"04",
                                                "5":"05",
                                                "6":"06",
                                                "7":"07",
                                                "8":"08",
                                                "9":"09"})
        distortionelbow(df,lections[i])
plt.title("Coude de distortion pour chaque lection")
plt.legend()
plt.savefig("figs/elbows.png")
plt.show()

for i, f in enumerate(glob.glob("./data_clean/*")):
    if i>=0:
        df = pd.read_csv(f, index_col=0).fillna(0)
        df.index = df.index.to_series().replace({"ZA": "971", 
                                                "ZB": "972", 
                                                "ZC": "973", 
                                                "ZD": "974", 
                                                "ZS": "975", 
                                                "ZP": "987", 
                                                "ZN": "988", 
                                                "ZM": "976", #"2A": 20, "2B": 20,
                                                "ZW": "986",
                                                "ZX": "978",
                                                "ZZ": "99",
                                                "1":"01",
                                                "2":"02",
                                                "3":"03",
                                                "4":"04",
                                                "5":"05",
                                                "6":"06",
                                                "7":"07",
                                                "8":"08",
                                                "9":"09"})
        fulldf.append(kmeans_elections(df, lections[i], 4))
```

```python
depdict = {"ZA": "971", 
            "ZB": "972", 
            "ZC": "973", 
            "ZD": "974", 
            "ZS": "975", 
            "ZP": "987", 
            "ZN": "988", 
            "ZM": "976", #"2A": 20, "2B": 20,
            "ZW": "986",
            "ZX": "978",
            "ZZ": "99",
            "1":"01",
            "2":"02",
            "3":"03",
            "4":"04",
            "5":"05",
            "6":"06",
            "7":"07",
            "8":"08",
            "9":"09"}

test = pd.DataFrame()
for i,l in enumerate(fulldf):
    l.index = list(map(lambda x: depdict.get(str(x), str(x)), l.index.tolist()))
    test = test.join(l,rsuffix=i,how="outer")

test.head()
```

```python
test.columns = lections
test.head()
```

```python
test = test.dropna()
test.head()
```

```python
from sklearn.metrics import v_measure_score

v_measures = np.zeros((len(test.columns),len(test.columns)))

for i, col in enumerate(test.columns):
    for j, col2 in enumerate(test.columns):
        v_measures[i,j] = v_measure_score(test[col],test[col2])

df_v = pd.DataFrame(v_measures)
df_v.index = test.columns
df_v.columns = test.columns
df_v.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_v,annot=True,fmt=".2f")
plt.suptitle("V_measure of kmean labels matrix",fontsize=30)
plt.savefig("figs/vmeasures.png")
plt.show()
```

```python
otherdict = {k:v for k in df_v.columns for v in df_v.columns if v!=k and v[:3]==k[:3]}
otherdict
```

```python
nearlist = ["Reg_15","Reg_21","Muni_14","Muni_20","Pdt_12","Pdt_17","Eur_14","Eur_19","Leg_12","Leg_17","Dep_15","Dep_21"]
neardict = {k:v for k,v in zip(df_v.columns, nearlist)}
neardict
```

```python
df_compare = pd.DataFrame.from_dict(otherdict, orient="index")
df_compare.columns = ["Same_election"]
df_compare["Nearest_election"] = neardict.values()
df_compare = df_compare.reset_index()
df_compare["Same_v_measure"] = df_compare.apply(lambda x: df_v.loc[x[0],x[1]],axis=1)
df_compare["Nearest_v_measure"] = df_compare.apply(lambda x: df_v.loc[x[0],x[2]],axis=1)
df_compare["Election"] = df_compare["index"]
df_compare = df_compare.drop(["index"],axis=1)
df_compare = df_compare.set_index("Election",drop=True)
df_compare
```
